```python
import math
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def black_scholes(S, K, T, r, sigma, option_type):
    d1 = (math.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)

    if option_type == 'call':
        option_price = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
        variance_vega = S * norm.pdf(d1) * math.sqrt(T)
    elif option_type == 'put':
        option_price = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
        variance_vega = S * norm.pdf(d1) * math.sqrt(T)

    return option_price, variance_vega

T = 1.0  
r = 0.05  
sigma = 0.2  
option_type = 'call'  

S_range = np.linspace(20, 200, 200)

K_values = np.linspace(30, 230, 20)

variance_vega_values = []
variance_vega_portfolio = np.zeros_like(S_range)
colors = cm.Blues(np.linspace(0.2, 1, len(K_values))) 
for i, K in enumerate(K_values):
    variance_vega_list = []
    for j, S in enumerate(S_range):
        _, variance_vega = black_scholes(S, K, T, r, sigma, option_type)
        variance_vega_list.append(variance_vega * (1 / K**2))
        variance_vega_portfolio[j] += variance_vega * (1 / K**2)
    variance_vega_values.append(variance_vega_list)
    plt.plot(S_range, variance_vega_list, color=colors[i])

plt.plot(S_range, variance_vega_portfolio, linestyle='--', color='black', label='Portfolio')

plt.xlabel('Underlying Asset Price')
plt.ylabel('Variance Vega')
plt.title('')
plt.yticks([])
plt.grid(True)
plt.legend()
plt.show()


```

```python
def black_scholes(S, K, T, r, sigma, option_type):
    d1 = (math.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * math.sqrt(T))
    d2 = d1 - sigma * math.sqrt(T)

    if option_type == 'call':
        option_price = S * norm.cdf(d1) - K * math.exp(-r * T) * norm.cdf(d2)
        variance_vega = S * norm.pdf(d1) * math.sqrt(T)
    elif option_type == 'put':
        option_price = K * math.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
        variance_vega = S * norm.pdf(d1) * math.sqrt(T)

    return option_price, variance_vega

# Parameters
T = 1.0  # Time to expiration in years
r = 0.05  # Risk-free interest rate
sigma = 0.2  # Volatility of the underlying asset
option_type = 'call'  # 'call' or 'put'

# Range of underlying asset prices (S)
S_range = np.linspace(20, 200, 200)

# List of strike prices (K)
K_values = np.linspace(0.1, 300, 40)

# Calculate variance vega for each underlying asset price (S)
variance_vega_values = []
variance_vega_portfolio = np.zeros_like(S_range)
colors = cm.Blues(np.linspace(0.2, 1, len(K_values)))  # Blue color gradient
for i, K in enumerate(K_values):
    variance_vega_list = []
    for j, S in enumerate(S_range):
        _, variance_vega = black_scholes(S, K, T, r, sigma, option_type)
        variance_vega_list.append(variance_vega * (1 / K**2))
        variance_vega_portfolio[j] += variance_vega * (1 / K**2)
    variance_vega_values.append(variance_vega_list)
    plt.plot(S_range, variance_vega_list, color=colors[i])

# Plot portfolio variance vega
plt.plot(S_range, variance_vega_portfolio, linestyle='--', color='black', label='Portfolio')

# Plot settings
plt.xlabel('Underlying Asset Price')
plt.ylabel('Variance Vega')
plt.title('')
plt.yticks([])
plt.grid(True)
plt.legend()
plt.show()
```
